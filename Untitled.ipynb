```python
def s(n):
    if n==0:
        return n
    else:
        return n+ s(n-1)
    
print(s(5))
```

```python
def L(n):
    if n==0:
        return 10
    else:
        return 5+ L(n-1)

print(L(1))
print(L(2))
print(L(3))
```

```python
def harmonic_sum(n):
    if n<=1:
        return 1
    else:
        return (1/n) + harmonic_sum(n-1)

print(harmonic_sum(5))
```

```python
def bacteria(n):
    if n == 1:
        return 2
    elif n == 2:
        return 3
    else:
        return bacteria(n - 1) * bacteria(n - 2)

print(bacteria(5))
```

```python
def pascal(n,k):
    if k==0:
        return 1
    elif n==k:
        return 1
    else :
        return pascal(n-1,k-1)+pascal(n-1,k)

print(pascal(4,2))
```

```python
def gcd(a, b):
    if b == 0:
        return a
    else:
        return gcd(b, a % b)

print(gcd(48, 18))
```

```python
def sum_even(n):
    if n < 2:
        return 0
    elif n % 2 != 0:
        return sum_even(n - 1)    
    return n + sum_even(n - 2)

print(sum_even(8))
```

```python
def savings(n):
    if n == 1:
        return 1000.0
    elif n == 2:
        return 1050.0
    else:
        return (1.05 * savings(n - 1)) + savings(n - 2)

print(savings(3))
```

```python
def factorial(k):
    if k <= 1:
        return 1
    return k * factorial(k - 1)

def sum_factorials(n):
    if n == 1:
        return 1
    return factorial(n) + sum_factorials(n - 1)

print(sum_factorials(4))
```
